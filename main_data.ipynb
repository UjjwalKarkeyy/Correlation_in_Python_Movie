## üîß Importing Libraries
The required Python libraries for data manipulation and visualization are imported here.

````python
# Import essential libraries
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

from matplotlib.pyplot import figure
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusting the figure size for the plots we create
````

## üìÇ Loading the Dataset
The movie dataset is read from a CSV file into a pandas DataFrame.

````python
# Read the dataset
# Reading the data
df = pd.read_csv('data/movies.csv')
````

## üëÄ Previewing the Data
Display the first few rows of the dataset to get an initial understanding of its structure.

````python
# Display the first 5 rows
df.head()
````

## üß± Viewing Column Names
Check the list of all column names in the dataset.

````python
# List all column names
df.columns
````

## üß± Viewing Missing Values
Check the percentage of all missing values in the dataset.

````python
# List all column names
# Checking for null values

for col in df.columns:
    missing =  np.mean(df[col].isnull())
    print("{} - {}%".format(col, missing))
````

## üß± Dropping missing rows

````python
df = df.dropna()
````

## üß± Viewing Missing Values Percentage after the drop

````python
# List all column names
# Checking for null values

for col in df.columns:
    # df[col].isnull() returns true or false i.., if we had [1,2,none,4,none] then it will return [false, false, true, false, true]
    # our np.mean will take those true/false as 1/0 and calculate the average i.e., [0,0,1,0,1] -> (0 + 0 + 1 + 0 + 1) / 5 -> 0.4 = 4%
    
    missing =  np.mean(df[col].isnull())  
    print("{} - {}%".format(col, missing))
````

````python
# Checking data types

df.dtypes
````

````python
# Changing data type of budget and gross column

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
````

````python
df.dtypes
````

````python
# Creating correct year column

df['year_correct'] = df['released'].astype(str)
df['year_correct'] = df['year_correct'].str.extract('(\d{4})')
````

## üëÄ Previewing the Data
Display the first few rows of the dataset to get an initial understanding of its structure.

````python
# Display the first 5 rows
df.head()
````

````python
# Sorting data frame by gross column

df.sort_values(by=['gross'], inplace=False, ascending=False)
````

````python
pd.set_option('display.max_rows', 50)
````

````python
# Removing duplicates

df.drop_duplicates()
````

````python
# Scatter plot for budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title("Budget vs Gross Earning")
plt.xlabel("Budget")
plt.ylabel("Gross Earning")
plt.show()
````

````python
# Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color":'red'}, line_kws={"color":'green'})
````

````python
# Looking at correlation

df.corr(numeric_only=True, method='pearson') # Different methods for finding correlation: Pearson (default), kendall, spearman
````

````python
corr_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Heatmap")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
````

````python
# Changing object type to numerical type

df_numerized = df.copy()

for col in df_numerized.columns:

    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
````
